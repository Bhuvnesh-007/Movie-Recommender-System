```python
# Code to merge these 2 dataframe
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
movies = movies[['movie_id','genres','keywords','cast','crew','overview','title']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres

# From this weird format we want a list format then a text type format
```

```python
import ast

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# Now we want only the top three cast of the Cast column

def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter < 3:
      L.append(i['name'])
      counter += 1
    else:
        break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
# Now we want the Director of movies only from the crew section

def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'].apply(lambda x:x.split())

# Converting the string into a list
```

```python
# Now we are going to remove the spaces in between any name

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies.apply(
    lambda row: ' '.join(
        row['overview'].split() +
        row['genres'] +
        row['keywords'] +
        row['cast'] +
        row['crew']
    ),
    axis=1
)
```

```python
movies = movies[['movie_id','title','tags']]
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['tags'].str.lower()
```

```python
movies.head(5)
```

```python
movies['tags'][0]
```

```python
# Now we are going to use the concept of vectorization

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Techique of Stemming

# Stemming means reducing words to their root/base form

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
movies['tags'].apply(stem)
```

```python
# We are going to find the similarity between the vectors using measure called cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors).astype(np.float32)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = movies[movies['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  
  for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Spaced Invaders')
```
